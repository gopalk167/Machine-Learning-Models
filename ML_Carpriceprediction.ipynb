## Cleaning of data

```python
backup = car.copy()
```

```python
car=car[car['year'].str.isnumeric()]
```

```python
car['year']=car['year'].astype(int)
```

```python
car=car[car['Price']!='Ask For Price']
```

```python
car['Price']=car['Price'].str.replace(",","").astype(int)
```

```python
car.info()
```

```python
car['kms_driven']=car['kms_driven'].str.split(" ").str.get(0).str.replace(",","")
```

```python
# car.info()
car=car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
car.shape
```

```python
car=car.dropna(axis=0)
```

```python
# keep 3 words of name
car['name']=car['name'].str.split(" ").str.slice(0,3).str.join(' ')
```

## Deploying  a linear Regression Model

```python
X= car.drop(columns ='Price')
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y , test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
column_trans= make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name',
                        'company','fuel_type']),remainder=
                                     'passthrough')
```

```python
column_trans
```

```python
lr=LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_predict=pipe.predict(X_test)
```

```python
y_predict
```

```python
r2_score(y_test,y_predict)
```

```python
score=[]
for i in range(1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y , test_size=.2, random_state=i)
    lr=LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_predict=pipe.predict(X_test)
    score.append(r2_score(y_test,y_predict))
score
```

```python
np.argmax(score)
```

```python
score[(np.argmax(score))]
```

```python
 X_train,X_test,y_train,y_test = train_test_split(X,y , test_size=.2, random_state=np.argmax(score))
lr=LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_predict=pipe.predict(X_test)
r2_score(y_test,y_predict)
```
